```python
data_path = 'data/'
opd_data_filename = data_path + 'OPD_demografia.xlsx'
opd_data_no_gps_filename = data_path + 'opd_Malaria_result5.csv'
```

```python
opd_data = pd.read_excel(opd_data_filename)
opd_data_no_gps = pd.read_csv(opd_data_no_gps_filename)
```

```python
np.random.seed(34686304)
opd_data = opd_data.sample(frac = 1)
```

```python
# Pre-processing OPD data
opd_data = geopandas.GeoDataFrame(opd_data, geometry = geopandas.points_from_xy(opd_data['longitude'], opd_data['latitude']))
opd_data = opd_data.set_crs(epsg=4326)
opd_data = opd_data.to_crs(epsg=32736)
time_diff = pd.to_datetime(opd_data['date']) - opd_data['dob']
opd_data['age'] = time_diff.astype('timedelta64[Y]')
opd_data['year'] = pd.to_datetime(opd_data['date']).dt.year
opd_data['visits'] = np.ones_like(opd_data['rdt'])
opd_data['testdone'] = np.array((opd_data['rdt'] == 1)|(opd_data['rdt'] == 2), dtype = int)
opd_data = pd.merge(opd_data, opd_data_no_gps[['serialno', 'temp', 'fever2yno', 'place']], on = 'serialno', how = 'left')
malaria_positive = ((opd_data['temp'] >= 37.5) | (opd_data['fever2yno'] == 1)) & \
                    ((opd_data['rdt'] == 1) | ((opd_data['thicksmear'] >= 1)&(opd_data['thicksmear'] <= 5)))
opd_data['malaria'] = np.array(malaria_positive, dtype = int)
place_dict = {
            'Manhiça' : 1,
            'Maragra' : 2,
            'Ilha Josina' : 3,
            'Taninga' : 4,
            'Palmeiras' : 5,
            'Malavele' : 6,
            'Xinavane' : 7
            }
for p in place_dict:
    mask = opd_data['place'] == place_dict[p]
    opd_data['place'][mask] = p
opd_data['date'] = pd.to_datetime(opd_data['date'])
```

```python
mask_geo = opd_data['latitude'].notnull() & opd_data['longitude'].notnull()
mask_date = (opd_data['date'] >= '11-2016') & (opd_data['date'] <= '11-2019')
mask_area = (opd_data['place'] == 'Ilha Josina')|(opd_data['place'] == 'Manhiça')
mask_age = opd_data['age'] < 5
mask_age210 = (opd_data['age'] >= 2)&(opd_data['age'] < 10)
opd_mask = mask_date&mask_geo&mask_age&mask_area
```

```python
opd_data['x'] = opd_data['geometry'][mask_geo].apply(lambda p: p.x)
opd_data['y'] = opd_data['geometry'][mask_geo].apply(lambda p: p.y)
```

### Comparing time and spatial location of hotspots

```python
def get_hotspot_positions(data, mask, testres, dates, fof_cat_test_epi, time_width, time_steps, scale, min_num, \
                           linking_time, linking_dist, min_date, max_date, date_name = 'visdate', label2plot = 'mean_pr', \
                          show_maps = False, gif_delay = 20, method = 'fof', \
                           output_path = '', save = False, \
                          max_p_lifetimes = 0.05, xlims = None, ylims = None, \
                           version = 'epifriends'):
    step_num = 0
    x_positions = np.array([])
    y_positions = np.array([])
    tempids = np.array([])
    
    while min_date + pd.to_timedelta(time_steps*step_num + time_width, unit = 'D') < max_date:
        #select data in time window
        selected_data = (dates >= min_date + pd.to_timedelta(time_steps*step_num, unit = 'D'))& \
                        (dates <= min_date + pd.to_timedelta(time_steps*step_num + time_width, unit = 'D'))&\
                        mask
        
        #Find indeces from outbreaks
        in_outbreaks = np.array(fof_cat_test_epi[step_num]['indeces'])
        
        
        if len(fof_cat_test_epi[step_num]) >= 1:
            #epi_catalogue_list[step_num].plot(ax = ax, facecolor = "None", \
            #                                  markersize = markersize*epi_catalogue_list[step_num]['total'], \
            #                                  alpha = .5, edgecolor = 'r', lw = 3)
            for ii, i in enumerate(in_outbreaks):
                x_positions = np.concatenate((x_positions, data[selected_data].iloc[i]['x']))
                y_positions = np.concatenate((y_positions, data[selected_data].iloc[i]['y']))
                tempids = np.concatenate((tempids, fof_cat_test_epi[step_num].iloc[ii]['tempID']*np.ones_like(data[selected_data].iloc[i]['y'])))
                #data[selected_data].iloc[i].plot(ax = ax, column = data[date_name][selected_data].iloc[i].view(int), \
                #                              markersize = markersize, alpha = 1, legend = False, \
                #                             edgecolor = 'tab:red', vmin = min_date.value, vmax = max_date.value, lw = 1)
        
    
        step_num += 1
    return x_positions, y_positions, tempids
```
